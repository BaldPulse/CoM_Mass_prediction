```python
# load log/log_train.txt

import numpy as np
import matplotlib.pyplot as plt

# load the log file
log_file = 'log/log_train.txt'
```

```python
# read each epoch's last batch's loss
loss = []
eval_loss = []
with open(log_file, 'r') as f:
    for line in f:
        if 'batch: 200' in line:
            loss.append(float(line.split()[-1]))
        if 'eval loss:' in line:
            eval_loss.append(float(line.split()[-1]))
```

```python
# plot the loss and eval loss
loss_epoch = range(len(loss))
plt.plot(loss_epoch, loss, 'b', label='loss')
eval_epoch = range(0, len(loss), 10)
plt.plot(eval_epoch, eval_loss, 'r', label='eval loss')
# title and labels
plt.title('Loss and Eval Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```
